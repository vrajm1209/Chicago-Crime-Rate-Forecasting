```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import random
import seaborn as sns
from fbprophet import Prophet
```

```python
#read the datasets
df1 = pd.read_csv('Chicago_Crimes_2005_to_2007.csv', error_bad_lines=False)
df2 = pd.read_csv('Chicago_Crimes_2008_to_2011.csv', error_bad_lines=False)
df3 = pd.read_csv('Chicago_Crimes_2012_to_2017.csv', error_bad_lines=False)
```

```python
df = pd.concat([df1, df2, df3], ignore_index=False, axis=0)
```

## Preparing data as needed by Facebook Prophet

```python
chicago_df = df.resample('D').size().reset_index()
```

```python
chicago_df
```

```python
#prophet needs the columns named as ds and y where ds is the date and y is the target label
chicago_df = chicago_df.rename(columns={'Date':'ds', 0:'y'})
```

```python
chicago_df
```

## Make Predictions

```python
m = Prophet()
m.fit(chicago_df)
```

```python
# forcasting into the future (365 days)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
forecast
```

### Plotting the forecast

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Crime Rate')
```

```python
figure2 = m.plot_components(forecast)
```
